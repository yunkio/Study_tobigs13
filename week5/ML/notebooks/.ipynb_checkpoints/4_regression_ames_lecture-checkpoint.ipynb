```python
def column_null_ratio(df, top_num):
    df = (all_data.isnull().sum() / len(all_data)) * 100
    df = df.drop(df[df == 0].index).sort_values(ascending=False)[:top_num]
    missing_data = pd.DataFrame({'Missing Ratio (%)' : df})
    display(missing_data.head(top_num))
    
def check_data(data):        
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        print('--- column headers ----')
        print(data.columns.values) 
        print('-----------------------')

        print('-- column data types --')
        print(data.dtypes)
        print('-----------------------')

        print('-- column statistics --')
        print(data.describe()) 
        print('-----------------------')

        print('-- column null count --')
        print(data.isnull().sum())
        print('-----------------------')
        
        print('-- column null ratio --')
        print(column_null_ratio(data, 10))
        print('-----------------------')
```

### 3.1.결측치 처리 Data Imputation

#### 3.1.2 휴리스틱(Heuristic)으로 결측치 처리

```python
print(all_data['PoolQC'].dtype)
print(all_data['PoolQC'].unique())
```

- Data description에 따르면 결측치는 수영장Pool이 없음을 의미
- 상식적으로 대다수 집에 수영장이 없는 것이 타당함

```python
all_data["PoolQC"] = all_data["PoolQC"].fillna("None") # 'None'은 Pool이 '없음'을 의미
print(all_data['PoolQC'].unique())
```

```python
## 비슷한 상황의 설명변수들 동일하게 처리
for col in ("MiscFeature", "Alley", "Fence", "FireplaceQu"):
    print(all_data[col].dtype)

for col in ("MiscFeature", "Alley", "Fence", "FireplaceQu"):
    all_data[col] = all_data[col].fillna("None")

print('\n')
    
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    print(all_data[col].dtype)
    
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    all_data[col] = all_data[col].fillna("None")

print('\n')

for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    print(all_data[col].dtype)

for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    all_data[col] = all_data[col].fillna('None')

print('\n')
    
print(all_data["MasVnrType"].dtype)

all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")

all_data['MSSubClass'] = all_data['MSSubClass'].fillna("None")
```

```python
## 역시, Data description에 맞춰서
all_data["Functional"] = all_data["Functional"].fillna("Typical")
```

```python
## 숫자형 변수에 대해서는 document 내용 확인하고 0 넣기
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    print(all_data[col].dtype)

for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    all_data[col] = all_data[col].fillna(0)

print('\n')    
    
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    print(all_data[col].dtype)

for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    all_data[col] = all_data[col].fillna(0)

print('\n') 

print(all_data["MasVnrArea"].dtype)    
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
```

#### 3.1.3. 중심위치 척도 (평균mean/중앙값median/최빈값mode)으로 결측치 처리

- 부지 관련 설명변수 LotFrontage의 경우, 같은 지역(neighborhood) 내에서는 비슷한 값 가질 가능성이 높음 

```python
# all_data.groupby("Neighborhood")["LotFrontage"].median()  # 지역별 부지 중앙값 확인

all_data["LotFrontage"] = \
all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median())) # Neighborhood별 LotFrontage 값들 묶은 object 활용
```

- 열 내에 RL 값이 압조적으로 많으므로 최빈값 사용

```python
print('ratio of RL counts:', all_data['MSZoning'].str.count('RL').sum()/all_data['MSZoning'].size)

all_data['MSZoning'] = all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0])
```

```python
## 비슷한 상황의 설명변수들 동일하게 처리
all_data['Electrical'] = all_data['Electrical'].fillna(all_data['Electrical'].mode()[0])

all_data['KitchenQual'] = all_data['KitchenQual'].fillna(all_data['KitchenQual'].mode()[0])

all_data['Exterior1st'] = all_data['Exterior1st'].fillna(all_data['Exterior1st'].mode()[0])
all_data['Exterior2nd'] = all_data['Exterior2nd'].fillna(all_data['Exterior2nd'].mode()[0])

all_data['SaleType'] = all_data['SaleType'].fillna(all_data['SaleType'].mode()[0])
```

Three of them are true outliers (Partial Sales that likely don’t represent actual market values) and two of them are simply unusual sales (very large houses priced relatively appropriately)."

```python
## GrLiveArea열 vs SalePrice열 상의 이상치 확인
plt.scatter(x = train['GrLivArea'], y = train['SalePrice'])
plt.vlines(4000, 0, 800000)
plt.text(4500, 400000, 'outliers', fontsize = 20)

plt.xlim((0, 6000))
plt.ylim((0, 800000))
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
# 우측 하단 이상치 확인
```

- 우측 하단의 이상치는 매우 높은 거실 크기/거주 면적에 불구하고 매우 낮은 가격을 나타냄. 이상치로 판단하고 제거 

```python
# logical indexing으로 이상치 특정함
train[train['GrLivArea']>4000]
```

```python
train = train.drop(train[train['GrLivArea']>4000].index)

plt.scatter(train['GrLivArea'], train['SalePrice'])
plt.text(4400, 400000, '(removed)', fontsize = 17)

plt.xlim((0, 6000))
plt.ylim((0, 800000))
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
```

## 4. 특성 엔지니어링

```python
## 숫자형 변수처럼 보이지만 실제로는 범주형 변수인 것들 변형
all_data['MSSubClass'] = all_data['MSSubClass'].apply(str)

all_data['OverallCond'] = all_data['OverallCond'].astype(str)

all_data['YrSold'] = all_data['YrSold'].astype(str)
all_data['MoSold'] = all_data['MoSold'].astype(str)
```

```python
## 라벨 인코딩
from sklearn.preprocessing import LabelEncoder
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold')
# process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[c].values)) 
    all_data[c] = lbl.transform(list(all_data[c].values))

# shape        
print('Shape all_data: {}'.format(all_data.shape))
```

```python
# 층별 넓이 더해서 전체 넓이 특성 생성
all_data['TotalSF'] = all_data['TotalBsmtSF'] + all_data['1stFlrSF'] + all_data['2ndFlrSF']
```

### 4.1 기타 등등: 가변수, skewed 특성의 분포 저장

#### Log transformation으로 종속변수 분포의 왜도 skewness 조정하기 

```python
sns.distplot(train['SalePrice'])

plt.ylabel('Frequency')
plt.title('SalePrice distribution')
plt.show()
```

```python
sns.distplot(train['SalePrice'] , fit=norm)

(mu, sigma) = norm.fit(train['SalePrice'])
print(mu, sigma)
plt.show()
```

- https://data.library.virginia.edu/understanding-q-q-plots/
- quantile-quantile plot help us assess if a set of data plausibly came from some theoretical distribution. 
- it's a scatterplot created by plotting two sets of quantiles against one another. If both sets of quantiles came from the same distribution, should see points forming a line, roughly straight.

```python
stats.probplot(train['SalePrice'], dist='norm', plot=plt)
plt.show()
```

```python
train["SalePrice"] = np.log1p(train["SalePrice"]) # applies log(1+x) to the column
# add 1 for small input values to avoid cases: log(0) = -inf by rounding error and increased accuracy

sns.distplot(train['SalePrice'] , fit=norm);

(mu, sigma) = norm.fit(train['SalePrice'])
print(mu, sigma)

plt.ylabel('Frequency')
plt.title('SalePrice distribution')

fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

- 종속변수 분포의 왜도가 줄어들고 가우시안 분포에 가까워졌다
- 향후, predict할 때 np.expm1 사용

```python
np.expm1(np.log1p(200000)) # 다시 값 복원하는 방법
```
