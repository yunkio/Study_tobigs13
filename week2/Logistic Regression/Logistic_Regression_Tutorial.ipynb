```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() # Scaling
scaler.fit_transform(df)
```

```python
df_new = pd.DataFrame(scaler.fit_transform(df), columns = cancer.feature_names)
```

```python
df_new
```

```python
target = pd.Series(cancer.target, dtype = "category")
df['class'] = target # class 0: malignant, class 1: benign
df.head()
```

```python
target.value_counts()
```

```python
# Scaled Data
df_new['class'] = target
```

```python
df_new.head()
```

### Logistic Regression
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) # class 예측
y_pred
```

```python
classifier.predict_proba(X_test) # The returned estimates for all classes are ordered by the label of classes.
```

```python
classifier.score(X_test, y_test) # return mean accuracy
```

```python
# Scaled Version

X_train2, X_test2, y_train2, y_test2 = train_test_split(df_new.iloc[:, :-1], df_new.iloc[:, -1], random_state = 0)
```

```python
classifier2 = LogisticRegression()
classifier2.fit(X_train2, y_train2)
```

```python
y_pred2 = classifier2.predict(X_test2) # class 예측
y_pred2
```

```python
classifier2.score(X_test2, y_test2) # Scaling 후 성능 향상 O.
```

### ROC curve를 그려보자
roc_curve의 두번째 인자값이 뭔지 생각해보시면 좋을 것 같습니다. 위에 걸어둔 사이킷런 문서 참고하세요!

```python
fpr, tpr, thresholds = roc_curve(y_test, classifier2.decision_function(X_test))
# 순서대로 fpr, tpr, thresholds를 리턴함!
# fpr = 1-specificity
# tpr = sensitivity
# thresholds = 해당 fpr, tpr을 리턴할 때의 thesholds(cut-off) 값!
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = fpr
y = tpr

plt.plot(x,y) # 간단하게 ROC 그려볼 수 있음.
plt.xlabel('1-Specificity')
plt.ylabel('Sensitivity')
```

```python
roc_auc_score(y_test, classifier2.decision_function(X_test)) # auc_score 계산
```

```python
# sklearn 패키지에서 아예 plot 그리는 메소드를 제공 (0.22.1로 업데이트해야 함)

plot_roc_curve(classifier2, X_test, y_test, response_method = 'decision_function')
```

```python
res = [[fpr_i, tpr_i, thres_i] for fpr_i, tpr_i, thres_i in zip(fpr, tpr, thresholds)]
```
