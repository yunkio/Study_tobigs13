: 3개의 범주형 변수에 대해 인코딩을 진행해보자.

```python
def class_encoding(x):
    if x == '강제':
        return 0
    else:
        return 1
```

```python
def bid_encoding(x):
    if x == '개별':
        return '0'
    elif x == '일괄':
        return '1'
    else:
        return '2'
```

```python
def do_encoding(x):
    if x == '부산':
        return 0
    else:
        return 1
```

```python
data.Auction_class = data.Auction_class.apply(class_encoding)
```

```python
data.Bid_class = data.Bid_class.apply(bid_encoding)
```

```python
data.addr_do = data.addr_do.apply(do_encoding)
```

: 여기서 우리는 bid_encoding을 할 때 Bid_class의 경우 3개의 인자로 되어있어 열을 3개로 바꿔줄 필요가 있다.

### VIF

```python
X = data.drop('Hammer_price', axis=1)
```

```python
y = data['Hammer_price']
```

```python
#VIF확인하기
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.sort_values(["VIF Factor"], ascending=[False])
```

: 역시 높은 수치의 VIF가 나온다. 높은 수치의 VIF를 제거해보고, 또한 상식적으로 알려져 있는 연관있는 변수를 제거해보자.
 - 예시) point_X, 와 point_Y는 위도와 경도로 우리는 앞서 서울과 부산의 데이터만 있으므로 없애보자.

## Task 2 - 2

#### 1) Beta 구하기

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_scaled.shape
```

```python
import numpy as np
from numpy.linalg import inv 

def estimate_beta(x, y):
    """구현해야 하는 부분"""
    beta_hat = np.dot(np.dot(inv(np.dot(x.T, x)), x.T), y)
    
    return beta_hat
```

```python
Beta0=np.ones((1905,1))
```

```python
X = np.concatenate((Beta0, X_scaled), axis=1)
```

```python
estimate_beta(X, y)
```

#### 2) MSE 구하기

```python
(X * estimate_beta(X,y)).shape
```

```python
(y - np.sum(X * estimate_beta(X,y), axis=1)).shape
```

```python
np.sum((y - np.sum(X * estimate_beta(X,y), axis=1))**2)
```
