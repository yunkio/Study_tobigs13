```python
## 모듈 불러오기
from matplotlib import image # Pillow works under the covers
from matplotlib import pyplot as plt

import pandas as pd
```

```python
## 사진 불러오기
data = image.imread(r'..\input\assorted\dog.png')
```

```python
print(data.dtype)
print(data.shape) # 4번째 채널은 opacity
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): # 이미지 - 배열형태         
    display(pd.DataFrame(data[:,:,0]))

plt.imshow(data) # 원본 이미지
plt.show()
```

```python
plt.imshow(data[:,:,0], cmap='gray') # [x, y, intensity] # R에 해당
plt.show()

plt.imshow(data[:,:,1], cmap='gray') # G에 해당
plt.show()

plt.imshow(data[:,:,2], cmap='gray') # B에 해당
plt.show()

plt.imshow(data[:,:,3], cmap='gray') # opacity # 카메라 사진은 최대 명료함(?) # 줄어들면 사진 투명(?)해짐
plt.show()
```

## 굳이 R,G,B 느낌 내기

```python
x1 = data.copy()
for i in range(x1.shape[0]):
    for j in range(x1.shape[1]):
        x1[i, j, 1] = 0
        x1[i, j, 2] = 0
```

```python
plt.imshow(x1)
```
