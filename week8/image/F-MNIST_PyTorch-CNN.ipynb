```python
## 보충. 추가 차원
print(np.ndarray(shape=(1,3,3), dtype=int)) # 28 대신에 3
print('\n')
print(np.ndarray(shape=(3,3), dtype=int))
```

```python
## 텐서 변환
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()

test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()

print(train_Y.shape)
print(test_X.shape)
print(test_Y.shape)
```

```python
train = TensorDataset(train_X, train_Y) # 피쳐변수와 타겟변수 텐서를 합침

print(train[0]) # 텐서의 첫 번째 데이터 역시 확인 가능
# 맨 끝에 tensor(9) 있음

train_loader = DataLoader(train, batch_size=100, shuffle=True) # 미니배치 분할
```

## 3. 모델 학습 및 선정

```python
# class Net(nn.Module):
#     def __init__(self):
#         super(Net, self).__init__()
#         # 합성곱층
#         self.conv1 = nn.Conv2d(1, 6, 5) # 입력 채널 수, 출력 채널 수, 필터 크기
#         self.conv2 = nn.Conv2d(6, 16, 5)
#         # 전결합층
#         self.fc1 = nn.Linear(256, 64) # 256 = (((28-5+1)/2 )-5+1)/2 * (((28-5+1)/2 )-5+1)/2 * 16
#         self.fc2 = nn.Linear(64, 10)
    
#     def forward(self, x):
#         x = F.max_pool2d(F.relu(self.conv1(x)), 2) # 풀링 영역 크기
#         x = F.max_pool2d(F.relu(self.conv2(x)), 2)
#         x = x.view(-1, 256)
#         x = F.relu(self.fc1(x))
#         x = self.fc2(x)
#         return F.log_softmax(x) # softmax 통해서 확률 형태로 값 배출
    
# model = Net()
```

```python
# # sequential
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.architecture = nn.Sequential(
            # C:1, IH: 28, IW: 28
            nn.Conv2d(1, 6, 5), # C:1, FN:6, F(=FH=FW):5, S:1(default), P:0(default)
            # C:6, OH_conv:(IH+2P-FH)/S+1=(28+2*0-5)/1+1=24, OW:24(OH 동일)
            nn.ReLU(True),
            # (위 복사) C:6, IH:24, OW:24
            nn.MaxPool2d(2), # F(=FH=FW):2, S:F=2(default)
            # C:6, OH_pool:(IH-FH)/S+1=(24-2)/2+1=12, OW:12(OH 동일)
            
            nn.Conv2d(6, 16, 5), # C:6, FN:16, F(=FH=FW):5, S:1(default), P:0(default)
            nn.ReLU(True),
            nn.MaxPool2d(2),
            
            ################
            nn.Flatten(),
            ################ 여기부터 fully connected layers
            
            nn.Linear(256, 64),
            nn.ReLU(True),
            #BatchNorm1d(64)
            
            nn.Linear(64, 10),
            nn.ReLU(True)
        )
        
    def forward(self, x):
        x = self.architecture(x)
        return F.log_softmax(x) # 
    
model = Net()
```

```python
criterion = nn.CrossEntropyLoss() # 오차함수 객체

optimizer = optim.SGD(model.parameters(), lr=0.01) # 최적화 객체
```

```python
# 학습
for epoch in range(50): # 50회
    total_loss = 0
    for train_x, train_y in train_loader: # 분할해 둔 데이터를 꺼내옴
#         train_x, train_y = Variable(train_x), Variable(train_y) # 이제 불필요
        optimizer.zero_grad() # gradient 초기화
        output = model(train_x) # 순전파
        loss = criterion(output, train_y) # loss 계산
        loss.backward() # 역전파 계산
        optimizer.step() # 가중치 weight 업데이트 
        total_loss += loss.item() # 누적오차 계산 # n회 반복마다 바꾼다고 값의 order가 바뀌진 않네 # loss.item()는 loss의 스칼라값  # loss.data[0] 에러 야기
    # 5회 반복마다 누적 오차 출력
    if (epoch+1) % 5 == 0:
        print(epoch+1, total_loss)
```

```python
result = torch.max(model(test_X).data, 1)[1] # 출력이 0 혹은 1이 되게 함
accuracy = sum(test_Y.data.numpy() == result.numpy()) / len(test_Y.data.numpy()) # 모형의 정확도 측정

print(accuracy*100, '%') # 모형의 정확도 출력
# random baseline: 10%
```
