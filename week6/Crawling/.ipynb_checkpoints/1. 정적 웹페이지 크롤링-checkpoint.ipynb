```python
import requests
```

```python
url = 'https://movie.naver.com/movie/running/current.nhn'
```

```python
res = requests.get(url)
```

```python
res
```

```python
res.status_code
```

```python
res.content
```

```python
res.content.decode('utf8')
```

```python
html = res.text
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('dt[class=tit]')
```

```python
for tag in soup.select('dt[class=tit]'):
    print(tag.text.strip())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
##   : find()
html_ul = soup.find('ul')
print(html_ul.prettify())
```

```python
##   : find_all()
html_meta = soup.find_all('li')
for i in html_meta:
    print(i)
```

```python
##     : find_all(, limit=)
html_meta = soup.find_all('meta', limit=3)
for i in html_meta:
    print(i)
```

```python
## (class)  
html_class = soup.find_all(class_ ='tit')
for i in html_class:
    print(i)
    #print(i.text)
```

```python
## (id)  
html_id = soup.find_all(id = 'content')
for i in html_id:
    print(i)
```

```python
##  
```

```python
### class   "tit" !!
html_attr = soup.find_all(attrs = {'class':'tit'})
for i in html_attr:
    print(i)
```

```python
### id   "content" !!
html_attr = soup.find_all(attrs = {'id':'content'})
for i in html_attr:
    print(i)
```

```python
### type    !!
html_attr = soup.find_all(attrs={'type': True})
for i in html_attr:
    print(i)
```

```python
##  ()  
html_ul = soup.find('ul')
print(html_ul.text)
```

```python
#     a 

data = soup.find('ul', class_='lst_detail_t1')
data_a = ''
for tag in data.find_all('a'):
    print(tag.text)
    data_a += tag.text
```
