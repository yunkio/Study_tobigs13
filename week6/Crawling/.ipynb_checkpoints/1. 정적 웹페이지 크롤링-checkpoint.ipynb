```python
import requests
```

```python
url = 'https://movie.naver.com/movie/running/current.nhn'
```

```python
res = requests.get(url)
```

```python
html = res.text
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('dt[class=tit]')
```

```python
for tag in soup.select('dt[class=tit]'):
    print(tag.text.strip())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
## 하나만 검색 : find(‘태그명’)
html_ul = soup.find('ul')
print(html_ul.prettify())
```

```python
## 모두 검색 : find_all(‘태그명’)
html_meta = soup.find_all('li')
for i in html_meta:
    print(i)
```

```python
## 숫자 제한 해서 검색 : find_all(‘태그명’, limit=숫자)
html_meta = soup.find_all('meta', limit=3)
for i in html_meta:
    print(i)
```

```python
## 클래스(class) 로 검색
html_class = soup.find_all(class_ ='tit')
for i in html_class:
    print(i)
    #print(i.text)
```

```python
## 아이디(id) 로 검색
html_id = soup.find_all(id = 'content')
for i in html_id:
    print(i)
```

```python
## 속성으로 검색
```

```python
### class라는 속성이 값이 "tit"인 녀석들을!!
html_attr = soup.find_all(attrs = {'class':'tit'})
for i in html_attr:
    print(i)
```

```python
### id라는 속성의 값이 "content"인 녀석들을!!
html_attr = soup.find_all(attrs = {'id':'content'})
for i in html_attr:
    print(i)
```

```python
### type라는 속성의 값이 있는 녀석들을!!
html_attr = soup.find_all(attrs={'type': True})
for i in html_attr:
    print(i)
```

```python
## 태그의 텍스트(내용) 만 검색
html_ul = soup.find('ul')
print(html_ul.text)
```

```python
# 검색된 태그 내의 모든 a태그를 조회

data = soup.find('ul', class_='lst_detail_t1')
data_a = ''
for tag in data.find_all('a'):
    print(tag.text)
    data_a += tag.text
```
